# Seaborn Tutorial

Going through the Seaborn tutorial as a means of learning both Seaborn and Jupyter.


```python
# Import the required Python modules

import numpy as np # NumPy is the fundamental package for array computing with Python.
import pandas as pd # Powerful data structures for data analysis, time series, and statistics
import matplotlib.pyplot as plt # Python plotting package
import seaborn as sns # seaborn: statistical data visualization
sns.set(style="darkgrid")
```

```python
# Load the tips dataset comprising of (244) ID, bill, tip, sex, smoker, day, time and size
tips = sns.load_dataset("tips")

# Draw a scatterplot. 
# A Scatterplot is the default type in relplot but could also say kind = "scatter"
sns.relplot(x="total_bill", y="tip", data=tips);
```

```python
#The 2 dimensional plot can be made 3-dimensional by adding a colour to identify another variable
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
# You can even use a different marker style for each class
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker",
        data=tips);
```

```python
# Adding both hue and marker types to make two distinctions
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips);
#But this is hard to differentiate
```

```python
# If you base the hue on numeric data it is on a spectrum
sns.relplot(x="total_bill", y="tip", hue="size", data=tips);
```

```python
# Customise the colour palette
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.5,l=.75", data=tips);
```

```python
# Or change the size deoending on the numeric value
sns.relplot(x="total_bill", y="tip", size="size", data=tips);
```

```python
# You can specify the size range of the data points
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=tips);
```

## Emphasising Continuity with Line Plots ##

```python
# Here we start using pandas dataframes
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))

# Then the seaborn plot with a kind of "line"
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```

```python
# Another randomly generated x,y dataset using pandas
df = pd.DataFrame(np.random.randn(500, 2).cumsum(axis=0), columns=["x", "y"])
# Here we are knocking off the assumed relationship of y as a function of x
sns.relplot(x="x", y="y", sort=False, kind="line", data=df);
```

## Aggregation and representing uncertainty ##

```python
# More complex datasets will have multiple measurements for the same value of the x variable. 
# The default behavior in seaborn is to aggregate the multiple measurements at each x value 
# by plotting the mean and the 95% confidence interval around the mean:
fmri = sns.load_dataset("fmri")
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri);
```
